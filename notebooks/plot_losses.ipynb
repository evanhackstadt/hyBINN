```python
# Plot losses recorded from specified runs in experiments/
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Run 3

```python
df = pd.read_csv("../experiments/runs/binn_20260316_lr1e3_wd1e4/run3.csv",
                 index_col=0)
df.head()
```

```python
# convert to long format for seaborn plotting
dfm = df.melt(id_vars='epoch', var_name='type', value_name='loss')
dfm.head()
```

```python
sns.lineplot(dfm, x='epoch', y='loss', hue='type')
```

```python
sns.lineplot(df, x='epoch', y='cindexes')
```
